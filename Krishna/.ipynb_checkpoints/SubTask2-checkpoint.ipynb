# SUB-TASK 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("im_op_1.jpg")
img2 = cv2.imread("im_op_2.jpg")
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB), cmap = 'gray')
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB), cmap = 'gray')
```

# ARITHMETIC OPERATIONS


# Addition

```python
addn = cv2.add(img1, img2)

plt.imshow(cv2.cvtColor(addn, cv2.COLOR_BGR2RGB), cmap = 'gray')

#Not a good idea as most pixels become white (eg 200 + 215 = 255 as only 8 bits are permitted)
```

```python
addn_w = cv2.addWeighted(img1, 0.5, img2, 0.4, 0)
plt.imshow(cv2.cvtColor(addn_w, cv2.COLOR_BGR2RGB), cmap = 'gray')

# w = alpha * img1 + beta * img2 + gamma
```

# Subtraction

```python
sub1 = cv2.subtract(img1, img2)
plt.imshow(cv2.cvtColor(sub1, cv2.COLOR_BGR2RGB), cmap = 'gray')
```

```python
sub2 = cv2.subtract(img2, img1)
plt.imshow(cv2.cvtColor(sub2, cv2.COLOR_BGR2RGB), cmap = 'gray')
```

# BITWISE OPERATORS

```python
bw1 = cv2.imread("bw1.png", 0)
bw2 = cv2.imread("bw2.png", 0)
```

```python
plt.imshow(bw1, cmap = 'gray')
```

```python
plt.imshow(bw2, cmap='gray')
```

## AND

```python
and_img = bw1 & bw2

plt.imshow(and_img, cmap='gray')

# 0 & X = 0, 1 & 1 = 1
```

## OR

```python
or_img = bw1 | bw2

plt.imshow(or_img, cmap='gray')

# 1 | X = 1, 0 | 0 = 0
```

## NOT

```python
not_img = ~(bw2)

plt.imshow(not_img, cmap='gray')

# ~1 = 0, ~0 = 1
```

## XOR

```python
xor_img = bw1 ^ bw2

plt.imshow(xor_img, cmap='gray')

# X ^ X = 0, X ^ 0 = X, X ^ 1 = X'
```

## XNOR (complement of XOR)

```python
xnor_img = bw1 ^ ~(bw2)

plt.imshow(xnor_img, cmap='gray')

# X @ X = 1, X @ 0 = X', X @ 1 = X
```
