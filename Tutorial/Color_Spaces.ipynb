```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Apple.jpeg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray',gray)
cv.waitKey(0)
```

BGR TO HSV

```python
HSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('hsv', HSV)
cv.waitKey(0)
```

BGR to L*A*B

```python
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow('hsv', lab)
cv.waitKey(0)
```

CV2 is BGR coded while MATPLOTLIB is RGB coded.

```python
plt.imshow(img)
plt.show()
cv.imshow('hsv', img)
cv.waitKey(0)
```

BGR to RGB

```python
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.show()
cv.imshow('RGB', img)
cv.waitKey(0)
```
